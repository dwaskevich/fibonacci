```python
#set limit for number of fibonacci elements to plot
limit = 11
sequence = []
x = []
y = []
for n in range(1,limit+1):
    print(n, end=" ")
    x.append(n)
    y.append(fibonacci.fibo(n))
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y, marker='o')
# Add value labels
for i in range(len(x)):
    plt.text(x[i], y[i], str(y[i]), ha='right', va='baseline')
ax.set_title("Fibonacci Sequence")
plt.show()
```

```python
import matplotlib.patches as patches

# Create a figure and an axes
fig, ax = plt.subplots()

# Set the limits of the plot
ax.set_xlim(-10, 2*fibonacci.fibo(limit))
ax.set_ylim(-10, 2*fibonacci.fibo(limit))

# Create first two rectangle patches
rect = patches.Rectangle((0,0), y[0], y[0], linewidth=1, edgecolor='blue', facecolor='none')
# Add the rectangle to the axes
ax.add_patch(rect)
rect = patches.Rectangle((1,0), y[1], y[1], linewidth=1, edgecolor='black', facecolor='none')
# Add the rectangle to the axes
ax.add_patch(rect)

# Create rectangle patches corresponding to Fibonacci values
for i in range(2, limit):
    if x[i] % 2 == 0:
        rect = patches.Rectangle((y[i-1], 0), y[i], y[i], linewidth=1, edgecolor='red', facecolor='none')
        # Add the rectangle to the axes
        ax.add_patch(rect)
        # Label patch with Fibonacci value
        if i > 3:
            ax.text(y[i-1] + 0.5 * y[i], 0.5 * y[i], str(y[i]), ha='center', va='center')
    else:
        rect = patches.Rectangle((0, y[i-1]), y[i], y[i], linewidth=1, edgecolor='green', facecolor='none')
        # Add the rectangle to the axes
        ax.add_patch(rect)
        # Label patch with Fibonacci value
        if i > 3:
            ax.text(0.5 * y[i], y[i], str(y[i]), ha='center', va='center')

# Display the plot
plt.show()
```
